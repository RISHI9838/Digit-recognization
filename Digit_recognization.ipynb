```python
#Importing digits image Data from mnist dataset
from keras.datasets.mnist import load_data
# Import numpy and pandas library
import numpy as np
import pandas as pd
```

```python
 #Splittting the data for training and testing purpose,
# xtr contains the training data and xts contains the result of xtr
# ytr contains the testing data and yts contains the result of yts
(xtr,xts),(ytr,yts) = load_data()
```

```python
# To check the shape of xtr
xtr.shape
```

```python
# To get the shape of ytr
print(ytr.shape)
```

```python
xtr = xtr.reshape(60000,784)  # Reshape train data
xts = xts.reshape(60000,1)    # Reshape train data results
ytr = ytr.reshape(10000,784)  # Reshape test data
yts  = yts.reshape(10000,1)   # Reshape test data results
x = np.vstack((xtr,ytr))      # To convert the train data into vertical form
y = np.vstack((xts,yts))      # To the convert the test data array into vertical
# Printing to check the data
print(xtr.shape)
print(xts.shape)
print(ytr.shape)
print(yts.shape)
```

```python
# Importing train_test_split to split the data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y)
```

```python
# Using RandomForest because it provides higher accuracy through cross validation,
# it reduces overfitting problem in decision trees and also reduces the variance and therefore improves the accuracy.
# Importing the RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
# Creating model of sklearn.ensemble._forest.RandomForestClassifier type 
rmodel = RandomForestClassifier() 
rmodel.fit(xtrain,ytrain)   # Train that model based on data
```

```python
#Training And Testing Accuracy
print(rmodel.score(xtrain,ytrain))  # Accuracy for training data
print(rmodel.score(xtest,ytest))    # Accuracy of testing data
```

```python
pred = int(input('Enter the row value : '))   # Taking the row value from user which row's data to be test
print('Prediction value : ',rmodel.predict(xtest[pred].reshape(1,784)))     #Prediction From testing data
print('Actual Answer: ',ytest[pred])    # Actual value of image
# Import the matplotlib to plot the image
import matplotlib.pyplot as plt
plt.imshow(xtest[pred].reshape(28,28))
plt.grid()
plt.show()
```
